```python
from flask import Flask, request, jsonify
from PIL import Image
import base64
import io
from pyngrok import ngrok

app = Flask(__name__)

@app.route("/tryon", methods=["POST"])
def tryon():

    person = request.files["person"]
    cloth = request.files["cloth"]

    person_img = Image.open(person.stream)
    cloth_img = Image.open(cloth.stream)

    # TEMPORARY DEMO RESULT
    # (later we replace this with the AI model)

    cloth_resized = cloth_img.resize(person_img.size)

    blended = Image.blend(person_img.convert("RGB"), cloth_resized.convert("RGB"), alpha=0.3)

    buffer = io.BytesIO()
    blended.save(buffer, format="PNG")

    img_str = base64.b64encode(buffer.getvalue()).decode()

    return jsonify({
        "resultImage": "data:image/png;base64," + img_str
    })


public_url = ngrok.connect(5000)
print("Public API URL:", public_url)

app.run(port=5000)
```

```python
import torch
from diffusers import DiffusionPipeline
from diffusers.utils import load_image

# switch to "mps" for apple devices
pipe = DiffusionPipeline.from_pretrained("yisol/IDM-VTON", dtype=torch.bfloat16, device_map="cuda")

prompt = "Turn this cat into a dog"
input_image = load_image("https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/diffusers/cat.png")

image = pipe(image=input_image, prompt=prompt).images[0]
```
